```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
im = cv.imread('data/Lenna.png')
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)
hsv = cv.cvtColor(im, cv.COLOR_RGB2HSV)
h, s, v = cv.split(hsv)
```

```python
hsv2 = []
for i in [h,s,v] :
    hsv2.append(cv.equalizeHist(i))
```

```python
change_v = cv.merge([h, s, hsv2[2]])
change_sv = cv.merge([h, hsv2[1], hsv2[2]])
change_hsv = cv.merge(hsv2)
```

```python
im_list = [im]
for i in [change_v, change_sv, change_hsv] :
    tmp = cv.cvtColor(i, cv.COLOR_HSV2RGB)
    im_list.append(tmp)
```

```python
for idx_x in range(len(im_list)):
    im = cv.cvtColor(im_list[idx_x], cv.COLOR_RGB2HSV)
    
    h, s, v = cv.split(im)
    
    fig, axes = plt.subplots(1, 4, figsize=(25, 3))
    
    axes[0].imshow(im_list[idx_x])
    axes[1].hist(h.ravel(), 256, [0,256])
    axes[2].hist(s.ravel(), 256, [0,256])
    axes[3].hist(v.ravel(), 256, [0,256])
```
