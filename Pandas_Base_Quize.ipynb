## 문제3(10점)
보스턴 전체 데이터셋에서 수치상으로 5개의 값만 결측치임을 표여주는 표현식을 작성하라.

## 문제4(5점)

결측치가 들어있는 행으로만 이루어진 데이터프레임을 가리키는 boston_with_nan을 정의하라.

```python
# pass와 None을 각각 적절한 코드와 표현식으로 대체하라.

# pass
na_list = []
for i in boston.columns :
    if boston[i].isnull().any() :
        na_list.append(i)

na_df_list = []
for i in na_list :
    na_df_list.append(boston[boston[i].isnull()])
    
boston_with_nan = pd.DataFrame(na_df_list[0])
#boston_with_nan = None

assert boston_with_nan.shape == (5,13)
```

## 문제5(5점)
결측치가 들어있는 행을 제외한 데이터프레임을 가지키는 boston_without_nan을 정의하라.

```python
# pass와 None을 각각 적절한 코드와 표현식으로 대체하라.

#pass

boston_without_nan = boston.dropna()
# boston_without_nan = None

assert boston_without_nan.shape == (501, 13)
```

## 문제6(5점)
boston_with_nan이 가리키는 데이터프레임에 사용된 결측치(NaN)을 모두 해당 특성의 평균값(mean)으로 대체하라. 즉, NaN이 실제로 대체되어야 한다.(경고가 발생하더라도 무시한다.)

```python
# boston_with_nan 데이터프레임에 포함된 모든 결측치를 평균값으로 대체해야함
# pass를 적절한 코드로 대체할 것

# pass
for i in na_list :
    boston[i].fillna(boston[i].mean(), inplace=True)

# boston_with_nan 데이터프레임에 결측치가 없음을 입증해야 함.
# None 을 적절한 표현식으로 대체할 것.

na_num = boston.isnull().sum().sum()

assert na_num == False
```
